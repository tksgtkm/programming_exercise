```python
import tiktoken

tik_tokenizer = tiktoken.get_encoding("gpt2")

text = "Hello, world. Is this-- a test?"
```

```python
integers = tik_tokenizer.encode(text, allowed_special={"<|endoftext|>"})

print(integers)
```

```python
strings = tik_tokenizer.decode(integers)

print(strings)
```

```python
print(tik_tokenizer.n_vocab)
```
